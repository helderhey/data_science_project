### Import Libraries and Define Auxiliary Functions

## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# If you were unable to complete the previous lab correctly you can uncomment and load this csv

# df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/api/dataset_part_2.csv')

df.head(5)
```

```python
df.shape
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass is also important; it seems the more massive the payload, the less likely the first stage will return.


```python
df['PayloadMass'].mean()
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class",
            data=df, aspect=80/20, s=10, legend=False)
plt.xlabel(f'\nFlight Number', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Pay load Mass (kg)\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(np.arange(df['FlightNumber'].min() - 1,
           df['FlightNumber'].max() + 2, step=4), fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc="best", frameon=True, fontsize=20,
           title='class', title_fontsize=20)
plt.show()
```

We see that different launch sites have different success rates.  <code>CCAFS LC-40</code>, has a success rate of 60 %, while  <code>KSC LC-39A</code> and <code>VAFB SLC 4E</code> has a success rate of 77%.


Next, let's drill down to each site visualize its detailed launch records.


### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value

sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class",
            data=df, aspect=80/20, s=10, legend=False)
plt.xlabel(f'\nFlight Number', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Launch Site\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(np.arange(df['FlightNumber'].min() - 1,
           df['FlightNumber'].max() + 2, step=4), fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc="best", frameon=True, fontsize=20,
           title='class', title_fontsize=20)
plt.show()
```

<h3 align="justify">In this graph we can see that the launch site with the highest number of operations for Falcon 9 is the CCAFS SLC-40, followed by the KSC LC 39A and, finally, by the VAFB SLC 4E. We can also get a visual idea of the success rates of recovering of the first stage of the spaceship. The VAFB SLC 4E and the KSC LC 39A have similar success rates of close to 77%, while the CCAFS SLC-40 has a success rate of 60%.</h3>

### TASK 2: Visualize the relationship between Payload and Launch Site


We also want to observe if there is any relationship between launch sites and their payload mass.


```python
# Plot a scatter point chart with x axis to be Pay Load Mass (kg) and y axis to be the launch site, and hue to be the class value

sns.catplot(y="LaunchSite", x="PayloadMass", hue="Class",
            data=df, aspect=80/20, s=10, legend=False)
plt.xlabel(f'\nPayload Mass (kg)', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Launch Site\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc="best", frameon=True, fontsize=20,
           title='class', title_fontsize=20)
plt.show()
```

<h3 align="justify">The data show that at the CCAFS SLC 40 and VAFB SLC 4E launch sites an increase in the Falcon 9 first stage recovery success rate when the payload used in the operation is higher than the average of that used in all operations. At KSC LC 39A this success rate does not change much.</h3>

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column

rate_suc = pd.DataFrame({'Rate Success' : df.groupby('Orbit')['Class'].mean('Class')}).sort_values(by=['Rate Success']).reset_index()
rate_suc
```

```python
sns.set(rc={'figure.figsize': (20, 6)})
sns.barplot(x="Orbit", y="Rate Success", data=rate_suc)
plt.xlabel(f'\nOrbit', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Success Rate\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

<h3 align="justify">In this graph we can see which orbits have the highest success rate in Falcon 9's first stage recovery.
The number of operations performed in each orbit should be considered. For example, ES-L1, HEO, SO and GEO have the record of only one operation in the analyzed data. The orbits with the most operations are GTO (27), ISS (21) and VLEO (14).</h3>

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value

sns.catplot(y="Orbit", x="FlightNumber", hue="Class",
            data=df, aspect=80/20, s=10, legend=False)
plt.xlabel(f'\nFlight Number', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Orbit\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(np.arange(df['FlightNumber'].min() - 1,
           df['FlightNumber'].max() + 2, step=4), fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc="best", frameon=True, fontsize=20,
           title='class', title_fontsize=20)
plt.show()
```

<h3 align="justify">In the orbits with the highest number of operations (GTO, ISS and VLEO) there seems not be strong relationship between the number of the flights and the success or the failure of Falcon 9's first stage recovery.</h3>

### TASK  5: Visualize the relationship between Payload and Orbit type


Similarly, we can plot the Payload vs. Orbit scatter point charts to reveal the relationship between Payload and Orbit type


```python
# Plot a scatter point chart with x axis to be Payload and y axis to be the Orbit, and hue to be the class value

sns.catplot(y="Orbit", x="PayloadMass", hue="Class",
            data=df, aspect=80/20, s=10, legend=False)
plt.xlabel(f'\nPayload Mass (kg)', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Orbit\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc="best", frameon=True, fontsize=20,
           title='class', title_fontsize=20)
plt.show()
```

<h3 align="justify">As seen earlier, the Falcon 9's first stage recovery success rate is higher in operations performed with a larger payload than the overall average.
On the other hand, operations in ES-L1, HEO and SSO orbits, performed with a smaller payload, had a 100% success rate.</h3>

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
rate_suc_year = pd.DataFrame({'Rate Success' : df.groupby('Year')['Class'].mean('Class')}).sort_values(by=['Rate Success']).reset_index()
rate_suc_year
```

you can observe that the sucess rate since 2013 kept increasing till 2020


```python
sns.lineplot(data=rate_suc_year, x="Year", y="Rate Success", linewidth=7.0)

plt.xlabel(f'\nYear', fontsize=20, color='blue', fontweight='bold')
plt.ylabel(f'Rate Success\n', fontsize=20,
           color='blue', fontweight='bold')
plt.xticks(np.arange(df['Year'].min(),
           df['Year'].max() + 1, step=1), fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

<h3 align="justify">In this graph we can check that the rate success of the recover of the first stage of Falcon 9 increase after 2013.
</h3>

## Features Engineering


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# HINT: Use get_dummies() function on the categorical columns

features_one_hot = features[['Orbit', 'LaunchSite', 'LandingPad','Serial']]

features = pd.concat([features, pd.get_dummies(features_one_hot).astype(float)], axis=1)
features.head()
```
